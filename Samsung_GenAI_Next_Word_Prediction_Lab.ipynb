```python
# Imports

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

MODEL_NAME = "Qwen/Qwen2.5-0.5B-Instruct"

print("Loading model...")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME
)

print("Model Loaded Successfully!")
```

# Task 1 (TODO)

Create a function called **predict_next_words()**

Inputs:
- prompt
- top_k

Outputs:
- Top-k predicted next tokens with probabilities.

**Hint**
- Tokenize the prompt
- Run the model
- Take the logits of the last token
- Apply softmax
- Use torch.topk()


```python
# ==========================
# TODO 1
# Complete this function
# ==========================

def predict_next_words(prompt, top_k=5):

    inputs = tokenizer(prompt, return_tensors="pt")

    with torch.no_grad():
        outputs = model(**inputs)

    logits = outputs.logits[:, -1, :]

    probs = torch.softmax(logits, dim=-1)

    values, indices = torch.topk(probs, top_k)

    predictions = []

    for score, idx in zip(values[0], indices[0]):
        token = tokenizer.decode([idx]).strip()
        predictions.append((token, float(score)))

    return predictions
```

# Task 2 (TODO)

Test your function using prompts like:

- Artificial Intelligence will

```python
prompt = "Artificial Intelligence will"

predictions = predict_next_words(prompt)

print("Top 5 Predictions\n")

for i, (word, score) in enumerate(predictions,1):
    print(f"{i}. {word} ({score:.4f})")
```

```python
# ============================================================
# BONUS TASK: Build a Gradio Web App
# ============================================================

import gradio as gr

def predict(prompt):

    predictions = predict_next_words(prompt, top_k=5)

    output = ""

    for i, (word, score) in enumerate(predictions,1):
        output += f"{i}. {word} ({score:.4f})\n"

    return output


demo = gr.Interface(
    fn=predict,
    inputs=gr.Textbox(
        lines=2,
        label="Enter Prompt",
        placeholder="Artificial Intelligence will"
    ),
    outputs=gr.Textbox(
        label="Top 5 Next Word Predictions"
    ),
    title="🚀 Anil Samsung GenAI Lab",
    description="Transformer - Next Token Prediction",
    examples=[
        ["Artificial Intelligence will"],
        ["The future of AI"],
        ["Samsung is developing"],
        ["Cloud computing enables"],
        ["Data Science is"]
    ]
)

demo.launch(share=True)
```
