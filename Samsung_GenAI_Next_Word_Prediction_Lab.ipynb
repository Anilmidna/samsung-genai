```python
# Imports

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

MODEL_NAME = "distilgpt2"

print("Loading model...")

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(MODEL_NAME)

print("Model Loaded Successfully!")
```

# Task 1 (TODO)

Create a function called **predict_next_words()**

Inputs:
- prompt
- top_k

Outputs:
- Top-k predicted next tokens with probabilities.

**Hint**
- Tokenize the prompt
- Run the model
- Take the logits of the last token
- Apply softmax
- Use torch.topk()


```python
# ==========================
# TODO 1
# Complete this function
# ==========================

def predict_next_words(prompt, top_k=5):

    # TODO
    inputs = tokenizer(prompt, return_tensors="pt")

    # TODO
    outputs = model(**inputs)

    # TODO
    logits = outputs.logits[:, -1, :]

    # TODO
    probs = torch.softmax(logits, dim=-1)

    # TODO
    values, indices = torch.topk(probs, top_k)

    results = []

    for score, idx in zip(values[0], indices[0]):
        token = tokenizer.decode([idx]).strip()
        results.append((token, float(score)))

    return results
```

# Task 2 (TODO)

Test your function using prompts like:

- Artificial Intelligence will
- Samsung is developing
- Cloud Computing enables
- The future of India


```python
prompts = [
    "Artificial Intelligence will",
    "Samsung is developing",
    "Cloud Computing enables"
]

for p in prompts:
    print("="*60)
    print("Prompt:", p)

    preds = predict_next_words(p)

    for i, (token, score) in enumerate(preds,1):
        print(i, token, round(score,4))
```

# Task 3 (TODO)

Generate multiple tokens.

Complete the function below.

Hint:
- Encode prompt
- Call model.generate()
- Decode output


```python
# ==========================
# TODO 2
# Complete this function
# ==========================

def generate_text(prompt, max_new_tokens=20):

    inputs = tokenizer(prompt, return_tensors="pt")

    output = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=True,
        temperature=0.8
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
prompt = "The future of AI"

print(generate_text(prompt))
```
